```python
import re
import datetime

with open("skwiki-latest-pages-articles1.xml", encoding='utf8') as f:
    content = f.readlines()
    
pocet = 0
pocet_u = 0
pocet_umr = 0

infobox = 0   #premenna na zistenie ci riadok sa nachadza v infoboxe alebo nie
for line in content:
    #prvy IF - zistujem ci premenna je nastavena na 1, co znamena ze riadok je z nfoboxu a treba ho zapisat
        if infobox == 1:
            f_infobox = open("infobox.txt", "a", encoding='utf8')
            f_infobox.write(line)
            f.close()
        
    #druhy IF - zistujem zaciatok infoboxu
        if re.search("{{Infobox Osobnosť", line):
            if re.search("{{Infobox Osobnosť}}", line):
                infobox = 0
            else:
                infobox = 1
                pocet = pocet + 1
                f_infobox = open("infobox.txt", "a", encoding='utf8')
                f_infobox.write(line)
                f.close()
       
    #IF - zistujem koniec infoboxu
        if infobox == 1:
            if re.search("Dátum úmrtia", line) or re.search("dátum úmrtia", line):
                infobox = 0
                f_infobox = open("infobox.txt", "a", encoding='utf8')
                f_infobox.write("ending story")
                f_infobox.write("\n")
                f.close()
                pocet_u = pocet_u + 1
                pocet_umr = pocet_umr + 1
            else:
                if re.search("^(\|\}\}\n)|( \}\}\n)|(\}\}\n)", line):
                    infobox = 0
                    f_infobox = open("infobox.txt", "a", encoding='utf8')
                    f_infobox.write("ending story")
                    f_infobox.write("\n")
                    f.close()
                    pocet_u = pocet_u + 1
print(pocet)
print(pocet_u)
print(pocet_umr)
```
